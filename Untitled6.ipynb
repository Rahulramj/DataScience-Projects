```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating sample age data
np.random.seed(0)
ages = np.random.randint(18, 80, size=100)  # 100 random ages between 18 and 80
```

```python
plt.hist(ages, bins=10, edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data_2 = pd.read_csv('2.csv')
data_3 = pd.read_csv('3.csv')

filtered_data_2 = data_2.dropna(subset=['IncomeGroup'])

income_group_counts = filtered_data_2['IncomeGroup'].value_counts()

plt.figure(figsize=(10, 6))
income_group_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.show()
```
